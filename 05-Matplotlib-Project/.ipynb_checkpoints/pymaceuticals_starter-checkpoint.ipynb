## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method is the most straighforward, creating multiple series and putting them all together at the end.
summary_stats = merged_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].describe()
summary_stats
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

mean_data = merged_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
mean_data
```

```python
median_data = merged_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
median_data
```

```python
variance_data = merged_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
variance_data
```

```python
standard_dev_data = merged_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
standard_dev_data
```

```python
sem = merged_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
sem
```

```python
summary_stats_pd = pd.DataFrame({"Mean": mean_data, "Median": median_data, "Variance": variance_data, "Standard Deviation": standard_dev_data, "SEM": sem})
summary_stats_pd
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens. Calculate the IQR and quantitatively determine if there are any potential outliers. 
merged_data
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

df7 = merged_data.sort_values(["Drug Regimen", "Mouse ID", "Timepoint"])
df7
```

```python
df8 = df7.loc[df7["Timepoint"] == 45]
df8.reset_index()
```

```python
df9 = df8[df8['Drug Regimen'].isin(['Capomulin'])]
df9.reset_index()
```

```python
convert_df = df9.sort_values(["Tumor Volume (mm3)"]).reset_index()
convert_df = convert_df["Tumor Volume (mm3)"]
convert_df
```

```python
quartiles = convert_df.quantile([.25,.5,.75])

lower_quartile = quartiles[0.25]
upper_quartile = quartiles[0.75]

inner_quartile = upper_quartile - lower_quartile

print(f"The upper quartile of temperatures is:{upper_quartile}")
print(f"The lower quartile of temperatures is:{lower_quartile}")
print(f"The interquartile range of temperatures is:{inner_quartile}")
print(f"The the median of temperatures is:{quartiles[0.5]} ")

lower_bound = lower_quartile - (1.5*inner_quartile)
upper_bound = upper_quartile + (1.5*inner_quartile)

print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
```

```python
fig1, ax1 = plt.subplots()

ax1.set_title('Final Tumor Volume in Regimen: Capomulin')
ax1.set_ylabel('Final Tumor Volume (mm3)')

ax1.boxplot(convert_df)

plt.show()
```

```python
avg_tumor.plot(kind = 'scatter', x = 'Weight (g)', y = 'Average Volume', grid=False, figsize = (5, 5), title = 'Weight (g) versus Average Volume of Tumor')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

mouse_weight = avg_tumor.iloc[:,0]
vol = avg_tumor.iloc[:,1]

correlation = st.pearsonr(mouse_weight, vol)
print(f"The correlation between both factors is {round(correlation[0],4)}")

```

```python
x_values = avg_tumor['Weight (g)']
y_values = avg_tumor['Average Tumor Volume']

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)

regress_values = x_values * slope + intercept

line_eq = "y = " + str(round(slope,4)) + "x + " + str(round(intercept,4))

plt.scatter(x_values, y_values)

plt.plot(x_values,regress_values,"r-")

plt.annotate(line_eq,(6,10),fontsize=15,color="red")


plt.xlabel('Mouse Weight')
plt.ylabel('Average Tumor Volume')
plt.show()
```

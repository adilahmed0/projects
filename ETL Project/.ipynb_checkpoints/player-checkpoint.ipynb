```python
import pandas as pd
from sqlalchemy import create_engine
```

### Read Michael Jordan Career CSV File and convert to pandas dataframe

```python
jordan_file = "jordan_career.csv"
jordan_df = pd.read_csv(jordan_file)
jordan_df.head()
```

### Narrow down columns to: to Michael Jordan - age, player ID, field goal percentage, free throw percentage, total rebounds, assists, steals, block shots and points

```python
jordan_df = jordan_df[["age","player_id","fgp","ftp","trb","ast","stl","blk","pts"]]
jordan_df.head()
```

### Read LeBron James Career CSV File and convert to pandas dataframe

```python
lebron_file = "lebron_career.csv"
lebron_df = pd.read_csv(lebron_file)
lebron_df.head()
```

### Narrow down to columns to: LeBron James - age, player ID, field goal percentage, free throw percentage, total rebounds, assists, steals, block shots and points

```python
lebron_df = lebron_df[["age","player_id","fgp","ftp","trb","ast","stl","blk","pts"]]
lebron_df.head()
```

### Read Michael Jordan and LeBron James yearly salary csv file; convert to pandas dataframe

```python
salaries_file = "salaries.csv"
salaries_df = pd.read_csv(salaries_file)
salaries_df.head()
```

### Narrow down columns to: player ID, salary, season and team

```python
salaries_df = salaries_df[["player_id","salary","season","team"]]
salaries_df.head()
```

### Sort salaries dataframe by "player_id"

```python
# salaries_df = salaries_df.set_index('player_id')
salaries_df.head()
```

### Use "loc" feature to narrow down to LeBron James and Michael Jordan salaries only

```python
filter_salaries = salaries_df.loc[(salaries_df["player_id"] == "jamesle01") | (salaries_df["player_id"] == "jordami01"), :]
filter_salaries
```

### Merge jordan_df and salaries_df

```python
jordan_merged_df = pd.merge(jordan_df,salaries_df, on = "player_id")
jordan_merged_df.head()
```

### Merge lebron_df and salaries_df

```python
lebron_merged_df = pd.merge(lebron_df,salaries_df, on = "player_id")
lebron_merged_df.head()
```

### Merge Jordan and LeBron merged dataframes by using "pd.concat"

```python
players_df = pd.concat([jordan_merged_df, lebron_merged_df])
players_df
```

```python
players_df
```

### Sort and rename player dataframe for clarity

```python
player_df = players_df.rename({"age": "Age",
                   "player_id": "Player_ID",
                   "fgp": "Field_Goal_Percentage",
                   "ftp": "Free_Throw_Percentage",
                   "trb": "Total_Rebounds",
                   "ast": "Assists",
                   "stl": "Steals",
                   "blk": "Block_Shots",
                   "pts": "Points",
                   "salary": "Salary",
                   "season": "Season",
                   "team": "Team"}, axis = 1)
player_df.head()
```

```python
player_df.dtypes
```

### Generate output CSV file

```python
player_df.to_csv("player.csv")
```
